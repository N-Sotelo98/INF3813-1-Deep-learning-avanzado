```python
import os
import pandas as pd
```

```python
# Path to the folder containing CSV files
folder_path = r'../data/cmg_historico/horario'

# List to store individual DataFrames
dataframes = []

# Iterate through all CSV files in the folder
for file_name in os.listdir(folder_path):
    print(file_name)
    if file_name.endswith(".csv"):  # Process only CSV files
        file_path = os.path.join(folder_path, file_name)
        
        # Extract year and month from the filename
        year_month = file_name[:6]  # e.g., '202211'
        year = year_month[:4]  # First 4 characters: '2022'
        month = year_month[4:]  # Last 2 characters: '11'

        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path, sep=';', decimal=',')
        
        # Replace patterns dynamically
        df['fecha'] = df['fecha'].str.replace(f',0/Unix{year}', '/')
        df['fecha'] = df['fecha'].str.replace(f',0/{year_month},', '/1')
        df['fecha'] = df['fecha'].str.replace('/00/', f'/{month}/')
        
        # Filter and process the DataFrame
        df = df[df['hora'] != 25]
        df['fecha_hora'] = df['fecha'] + ' ' + (df['hora'] - 1).astype(str) + ':00:00'
        df['fecha_hora'] = pd.to_datetime(df['fecha_hora'], format='%d/%m/%Y %H:%M:%S')
        
        # Append to the list
        dataframes.append(df)

```

```python
# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)

combined_df
```

```python
combined_df.drop_duplicates(subset=['fecha_hora','barra'], keep='last', inplace=True)

combined_df
```

```python
combined_df.to_csv(r'../data/cmg_historico/cmg_all.csv', index=False)
```
